## Data Transformation

```python
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
onehot_encoder = preprocessing.OneHotEncoder()
```

```python
# cols that have various values
other_cols
```

```python
df[other_cols].head()
```

```python
# drop unnecessary variables
df = df.drop(['Gender','Region'], axis=1)
```

```python
# impute unknown value
df.loc[df['Change_Risk_Segment'] == 'Unknown', 'Change_Risk_Segment'] = 'No change'
df.loc[df['Change_T_Score'] == 'Unknown', 'Change_T_Score'] = 'No change'
```

```python
def label(data, feature):
    data[feature] = label_encoder.fit_transform(data[feature])
```

```python
def transform_speciality(value):
#     transform medical speciality
    if 'MEDICINE' in value.split(' '):
        return 'MEDICINE'
    elif 'SURGERY' in value.split(' '):
        return 'SURGERY'
    elif df['Ntm_Speciality'].value_counts()[value] < 10 or value == 'Unknown':
        return 'OTHER'
    return value
```

```python
df['Ntm_Speciality']
```

```python
# Transform NTM Speciality
df['Ntm_Speciality'] = df['Ntm_Speciality'].apply(lambda x: transform_speciality(x))
```

```python
# label encoding variables with only two values to 0 and 1
label(df, 'Ntm_Specialist_Flag')
label(df, 'Risk_Segment_Prior_Ntm')
label(df, 'Tscore_Bucket_Prior_Ntm')
label(df, 'Adherent_Flag')
```

```python
df['Race'].unique()
```

```python
df.loc[df['Ethnicity'] != 'Not Hispanic', 'Ethnicity'] = 'Other'
df.loc[df['Race'] != 'Caucasian', 'Race'] = 'Other'
```

```python
X = df.drop(['Persistency_Flag','Ptid', 'target'], axis=1)
y = df['target']
```

```python
models = [
    DummyClassifier(strategy = 'most_frequent', random_state = 1),SVC(gamma='auto'), NuSVC(gamma='auto'), LinearSVC(),
    SGDClassifier(), KNeighborsClassifier(),
    LogisticRegression(), LogisticRegressionCV(cv=5),
    DecisionTreeClassifier(),
    BaggingClassifier(), ExtraTreesClassifier(n_estimators=300),
    RandomForestClassifier(n_estimators=300),
    xgb.XGBClassifier(objective="binary:logistic", random_state=1)
]

def score_model(X, y, estimator, score, **kwargs):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

    model.fit(X_train, y_train, **kwargs)
    expected  = y_test
    predicted = model.predict(X_test)

    print("{}: {}".format(estimator.__class__.__name__, score(expected, predicted)))

warnings.filterwarnings('ignore')

for model in models:
    score_model(X, y, model, accuracy_score)
```

```python
# show confusion matrix numbers
def cf_graph(y_test,pred):
    cf_matrix = metrics.confusion_matrix(y_test, pred)
    group_names = ['True Pos','False Neg','False Pos', 'True Neg']
    group_counts = ["{0:0.0f}".format(value) for value in
                    cf_matrix.flatten()]
    group_percentages = ["{0:.2%}".format(value) for value in
                         cf_matrix.flatten()/np.sum(cf_matrix)]
    labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
              zip(group_names,group_counts, group_percentages)]
    labels = np.asarray(labels).reshape(2,2)
    sns.heatmap(cf_matrix, annot=labels,xticklabels= ['Persistent','Non-Persistent'], yticklabels= ['Persistent','Non-Persistent'], fmt='')
    plt.show()
```

## Extreme Gradient Boosting Model

```python
clf = xgb.XGBClassifier(objective="binary:logistic", random_state=1)
```

```python
from bayes_opt import BayesianOptimization
from sklearn.model_selection import cross_val_score
```

```python
def gbm_cl_bo(max_depth, learning_rate, n_estimators, subsample):
#     use bayesian optimization process hyperpaaramter tuning
#     return model evaluation score
    params_gbm = {}
    params_gbm['max_depth'] = round(max_depth)
    params_gbm['learning_rate'] = learning_rate
    params_gbm['n_estimators'] = round(n_estimators)
    params_gbm['subsample'] = subsample
    scores = cross_val_score(xgb.XGBClassifier(objective="binary:logistic",random_state=1, **params_gbm),
                             X_train, y_train, scoring='accuracy', cv=5).mean()
    score = scores.mean()
    return score
```

```python
params_gbm ={
    'max_depth':(1, 10),
    'learning_rate':(0.01, 1),
    'n_estimators':(80, 500),
    'subsample': (0.8, 1),
}
gbm_bo = BayesianOptimization(gbm_cl_bo, params_gbm, random_state=1)
gbm_bo.maximize(init_points=20, n_iter=4)
```

```python
params_gbm = gbm_bo.max['params']
params_gbm['max_depth'] = round(params_gbm['max_depth'])
params_gbm['n_estimators'] = round(params_gbm['n_estimators'])
params_gbm
```

```python
clf = xgb.XGBClassifier(objective="binary:logistic", random_state=1,learning_rate=0.29489758520048526,max_depth=2, 
                        n_estimators=88, subsample=0.9357671065879782)
```

```python
from matplotlib.ticker import FormatStrFormatter
```

```python
def plot_annot(plot):
    for bar in plots.patches:
   
  # Using Matplotlib's annotate function and
  # passing the coordinates where the annotation shall be done
  # x-coordinate: bar.get_x() + bar.get_width() / 2
  # y-coordinate: bar.get_height()
  # free space to be left to make graph pleasing: (0, 8)
  # ha and va stand for the horizontal and vertical alignment
        plots.annotate(format(bar.get_height(), '.5f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=10, xytext=(0, 8),
                   textcoords='offset points')
```
